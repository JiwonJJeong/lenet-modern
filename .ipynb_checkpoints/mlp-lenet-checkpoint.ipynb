# MLP to LeNet
This notebook will investigate insights and differences between two deep learning architectures: build from MLPs and CNNs. This focuses on the FashionMNIST dataset, an image dataset of various clothing items labeled by what they are. The images are relative simple, being 28x28 pixels, allowing for a image-based multiclass classification task.

```python
import torch
from torch import nn
from d2l import torch as d2l
```

## Data Visualization
The FashionMNIST dataset is loaded using many of D2L's functions, showing the power of object-oriented software design.

```python
data = d2l.FashionMNIST(batch_size=128)
```

```python
data.visualize(next(iter(data.val_dataloader())))
```

Each data sample is a 28 by 28 pixel image. There is only a single channel.

## Multilayer Perceptron
We start the baseline classification task with a multilayer perceptron. To start, we do a standard MLP on the dataset. In order to processes 2D image data, we must flatten out the images into a single dimension.

```python
class MLP(d2l.Module):
    def __init__(self, lr=0.1, num_classes=10, hidden_layers=(256,)):
        super().__init__()
        self.hidden = nn.Sequential(
            *(nn.Sequential(nn.LazyLinear(h), nn.ReLU()) for h in hidden_layers)
        )
        self.out = nn.LazyLinear(num_classes)
    
    def forward(self, x):
        hidden_out = self.hidden(x)
        return self.out(hidden_out)
```

```python
mlp = MLP(hidden_layers=(128,64))
mlp.state_dict()
```

```python
trainer = d2l.Trainer(max_epochs=10, num_gpus=1)
trainer.fit(mlp, data)
```

```python
class LeNet(d2l.Classifier):
    """This is the base lenet model (with some adjustment of activation function for computational ease)."""
    def __init__(self, lr=0.1, num_classes=10):
        super().__init__()
        self.save_hyperparameters()
        self.net = nn.Sequential(
            nn.LazyConv2d(6, kernel_size=5, padding=2), nn.Sigmoid(),
            nn.AvgPool2d(kernel_size=2, stride=2),
            nn.LazyConv2d(16, kernel_size=5), nn.Sigmoid(),
            nn.AvgPool2d(kernel_size=2, stride=2),
            nn.Flatten(),
            nn.LazyLinear(120), nn.Sigmoid(),
            nn.LazyLinear(84), nn.Sigmoid(),
            nn.LazyLinear(num_classes))
```

```python
def init_cnn(module):
    """Initialize weights for CNNs."""
    if type(module) == nn.Linear or type(module) == nn.Conv2d:
        nn.init.xavier_uniform_(module.weight)
```

```python
model = LeNet(lr=0.1)
model.apply_init([next(iter(data.get_dataloader(True)))[0]], init_cnn)
[(name, param.shape) for name, param in model.named_parameters()]
```

```python
model.state_dict()
```

```python
trainer = d2l.Trainer(max_epochs=10, num_gpus=1)
trainer.fit(model, data)
```
